```python

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('SM.csv')
X = dataset.iloc[:, :2].values
y1 = dataset.iloc[:, -3].values
y2 = dataset.iloc[:, -2].values
y3 = dataset.iloc[:, -1].values
x_res = [['SBI',593.85]]

# Encoding the Independent Variable
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
x_res = np.array(ct.transform(x_res))

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X[:, :-1] = sc.fit_transform(X[:, :-1])
x_res[:, :-1] = sc.transform(x_res[:, :-1])

# Training the Random Forest Regression model on the whole dataset
from sklearn.ensemble import RandomForestRegressor
rfregressor1 = RandomForestRegressor(n_estimators = 100, random_state = 0)
rfregressor2 = RandomForestRegressor(n_estimators = 100, random_state = 0)
rfregressor3 = RandomForestRegressor(n_estimators = 100, random_state = 0)
rfregressor1.fit(X, y1)
rfregressor2.fit(X, y2)
rfregressor3.fit(X, y3)

#Predicting new result
print("High")
print(rfregressor1.predict(x_res))
print("")
print("Low")
print(rfregressor2.predict(x_res))
print("")
print("Close")
print(rfregressor3.predict(x_res))
```

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('SM.csv')
X = dataset.iloc[:, :2].values
y1 = dataset.iloc[:, -3].values
y2 = dataset.iloc[:, -2].values
y3 = dataset.iloc[:, -1].values
x_res = [['SBI',599.45]]

# Encoding the Independent Variable
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
x_res = np.array(ct.transform(x_res))

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X[:, :-1] = sc.fit_transform(X[:, :-1])
x_res[:, :-1] = sc.transform(x_res[:, :-1])

# Training the Polynomial Regression model on the Training set
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly_reg1 = PolynomialFeatures(degree = 2)
X_poly1 = poly_reg1.fit_transform(X)
plregressor1 = LinearRegression()
plregressor1.fit(X_poly1, y1)
poly_reg2 = PolynomialFeatures(degree = 2)
X_poly2 = poly_reg2.fit_transform(X)
plregressor2 = LinearRegression()
plregressor2.fit(X_poly2, y2)
poly_reg3 = PolynomialFeatures(degree = 2)
X_poly3 = poly_reg3.fit_transform(X)
plregressor3 = LinearRegression()
plregressor3.fit(X_poly3, y3)

#Predicting new result
print("High")
print(plregressor1.predict(poly_reg1.transform(x_res)))
print("")
print("Low")
print(plregressor2.predict(poly_reg2.transform(x_res)))
print("")
print("Close")
print(plregressor3.predict(poly_reg3.transform(x_res)))

```
